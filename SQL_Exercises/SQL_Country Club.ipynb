```python
import sqlite3
import pandas as pd
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn


def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
#Q10
def Q10(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
    SELECT s.name, s.Revenue
    FROM(
    SELECT f.name, SUM(CASE WHEN b.memid=0 THEN b.slots*f.guestcost
    ELSE b.slots*f.membercost END) as Revenue
    FROM Facilities as f
    INNER JOIN
    Bookings as b
    ON f.facid=b.facid
    GROUP BY f.name) as s
    WHERE s.Revenue < 1000
    ORDER BY s.Revenue;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Q10. List of Facilities Ordered by Total Revenue with Revenue Lesser than 1000")
        Q10(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
#Q11
def Q11(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
    SELECT m.memid, m.firstname as Member_FirstName,m.surname as Member_SurName, o.surname || ',' || o.firstname as Recommender_Name
    FROM MEMBERS as m
    INNER JOIN 
    MEMBERS as o
    ON o.memid=m.recommendedby
    WHERE m.memid != 0
    ORDER BY o.surname, o.firstname
    ;
    
    """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def Q11_main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Q11. Report of members and who recommended them in alphabetic surname,firstname order")
        Q11(conn)
 
 
if __name__ == '__main__':
    Q11_main()
```

```python
#Q12
def Q12(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
    SELECT b.facid,f.name, SUM(b.slots)
    FROM BOOKINGS as b
    INNER JOIN 
    FACILITIES as f
    ON b.facid=f.facid
    WHERE b.memid != 0
    GROUP BY b.facid
    ;
    
    """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def Q12_main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Q12. Facilities with their usage by member, but not guests")
        Q12(conn)
 
 
if __name__ == '__main__':
    Q12_main()
```

```python
#Q13
def Q13(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
    SELECT b.facid,f.name, strftime('%m', b.starttime) as Month, SUM(b.slots) as Usage
    FROM BOOKINGS as b
    INNER JOIN 
    FACILITIES as f
    ON b.facid=f.facid
    WHERE b.memid != 0
    GROUP BY b.facid, Month
    ;
    
    """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def Q13_main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Q13. Facilities Usage by Month, but not Guests")
        Q13(conn)
 
 
if __name__ == '__main__':
    Q13_main()
```
